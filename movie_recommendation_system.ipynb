# Data Processing

```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv.zip')
credits = pd.read_csv('tmdb_5000_credits.csv.zip')

movies
# it is data of movies have all the details of movies

movies.columns
# these are all columns in this dataset

movies.shape
# have approx 4800 movies and 20 columns of data

credits
# it have data of title cast and crew

credits.shape
# have same no. of data and 4 cols
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')

# now here we have got the int type data so now our func will work
```

```python
def convert(obj):

    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name']) 
    return L
```

```python
movies['cast'].iloc[0]
# it will give us cast of first movie only 
# now here again we want only the name of cast not all other data 
# so we will again create a program
```

```python
def convertCast(obj):

    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])

    return L

# it will give us only name of cast
```

```python
movies.loc[:, 'cast'] = movies['cast'].apply(convertCast)
```

```python
movies.head()
# here we have only names of cast 
```

```python
# now we have to do the same for crew
# but here we want only director not others

# let create a function 
def convertDir(obj):

    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break

    return L
```

```python
movies.loc[:, 'crew'] = movies['crew'].apply(convertDir)
```

```python
movies.head()
# here we got only director name at place of crew
```

```python
# there was some problem in professors program when we are trying to do in newer version of pandas so we are creating a func which will remove spaces

def remove_spaces(lst):
    if isinstance(lst, list):
        return [i.replace(" ", "") for i in lst]
    return lst  # Return the value as is if it's not a list
```

```python
movies.loc[:, 'genres'] = movies['genres'].apply(remove_spaces)
movies.loc[:, 'keywords'] = movies['keywords'].apply(remove_spaces)
movies.loc[:, 'cast'] = movies['cast'].apply(remove_spaces)
movies.loc[:, 'crew'] = movies['crew'].apply(remove_spaces)
```

```python
# now lets creat a new column named tags which will have all the data form overview, genres, keywords, cast and crew

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# now we will remove all columns except movie_id, title, and tags

movies = movies[['movie_id', 'title', 'tags']]
movies.head()
# we got our data what we wanted
```

```python
# now lets make this as a single str as it have so many data in a list

movies.loc[:,'tags'] = movies['tags'].str.join(' ')
```

```python
# we want all the tags data in lower case

movies.loc[:, 'tags'] = movies['tags'].apply(lambda x: ' '.join(x).lower() if isinstance(x, list) else (x.lower() if isinstance(x, str) else x))
```

```python
# but all the data are text not int that we can calculate similarity
```

```python
# so now in this situation we have vectorization helps

# we will convert all the movies to a vector on the basis of count of data

# here we will to text vectorization  using : bag of words method
# in this method firstly we will concatinate every string and create a single string 
# new_tag = tag[0] + tag[1] + tag[2] + tag[3] + ....
# now here we will count down all the words frequency 10000 most common words
# now next we will ask every movies tag that how many times all words from that 10000 words are here in tags
# so this way for the first movies we have a vecotor having 10000 columns which are most common words
# now we will do this same thing for all 44400 movies 
# so we will get a  vector of size 44400 rows and 10000 columns

# one more important thing we will not consider stop words: stop words means words which are only helping for sentence creation
# like: is, am , are , the , and, too, a, of, and  so many
# we will apply vectorization on other words


# may be it seems very hard in listening but we have not to do something to much 
# sklearn will help us to do this so it will become very easy
# in sklearn we have a class from sklearn.feature_extraction.text named: CountVectorizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# now we will create a class of CountVectorizer
cv =  CountVectorizer(max_features = 5000, stop_words = 'english')
# here we have choose no of words to 5000 and using stop_words we are preventing to use sentence builder words from English
```

```python
# cv.fit_transform(movies['tags'])
# it returns us a sparse matrix so we will convert it to array

vect_arr = cv.fit_transform(movies['tags']).toarray()
# here we have an array having 5000 data in each row and total 4800 rows
```

```python
vect_arr.shape
```

```python
# we can see that what is our 5000 words using a method of cv

# cv.get_feature_names()
# it was using previously now its renamed
cv.get_feature_names_out()
# here we have 5000 most common words
```

```python
# we have use steming
# steming is a technique which is use for removing words having same meaning but spell is different

['love', 'loves', 'loving', 'loved']
# after steming we will get
['love', 'love', 'love', 'love']

# now we will apply steming on whole data 
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stemmer(text):

    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies.loc[:, 'tags'] = movies['tags'].apply(stemmer)

# how it will did stemming and we will get new data which is will be changed
```

```python
movies['tags']
```

```python
# now we will do that sklearn process again on this new data

cv =  CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vect_arr = cv.fit_transform(movies['tags']).toarray()
# here we have an array having 5000 data in each row and total 4800 rows
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vect_arr) 
```

```python
similarity[0]
# it is similarity score of first movie by all other movies
# it will show us firs item as 1 because first item is similarity with it self so similarity with it self is 1

# here we will sort this data than we will get top 5 movies which are most similar to the first movie
# if some one will ask for first movie than we will recommend them these 5 movies
```

```python
sorted(similarity[0], reverse = True)
# it is in ascending order we want in descending order
# but we will lost the index position because of sorting 
# so some how we have to stick with index
```

```python
# for that problem we will use enumerate fucntion 
# sorted(list(enumerate(similarity[0])))
# here sorting is done by in asceding order

# sorted(list(enumerate(similarity[0])), reverse = True)
# again sorting is done but it does both the data index and value 
# we want to sort only value data

sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
# now using this we will sort only the values not index
# so here second value is (2979, 0.48433244995346114),
# it means most similar movie with first is at index 2979
# we want 5 movies so we will use indexing for that
# here we have 5 movies which are most similar
```

## Recommend function which will give us movies 

```python
def recommend(movie):

    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]

    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    recommended_movie = []
    for i in movies_list:
        recommended_movie.append(movies.iloc[i[0]].title)
    return recommended_movie
```

```python
recommend('The Shawshank Redemption')

# here we have given a movie name in the fucntion and it will give us index of top 5  movies which are similar
# but it is giving us index we want name of movies
# now final model is ready 
# it can give us most similar movies 
```
